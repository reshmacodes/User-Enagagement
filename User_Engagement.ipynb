Analyzing User Engagemment Activity

### Question 1
Has user activity or engagement dropped, increased or remained stable? What is the extent of change in user activity/engagement?

#### Analysis

The data shows a gradual decrease in user engagement on Zillto Portal over the period of four month of data. There are no sharp peeks or falls are found during this time. It has also been found that the highest user engagement at the end of July is 8291 which eventually drops in the first week of August to 6425. Further analysis on action and location brings light to this sudden drop in engagement.
The extent of change is found to be 44.28% in the span of four months.

```
%sql
SELECT substring(date_trunc('week', occurred_at),1,10) as Activity, count(1) as User_Engagement
FROM events1
WHERE event_type = 'engagement'
group by 1
order by 1
```

```
%sql
Select ROUND(((MAX(User_Engagement)- MIN(User_Engagement))/MIN(User_Engagement))*100 ,2)as Percent_Extent_Change
from(

(SELECT substring(date_trunc('week', occurred_at),1,10) as Activity_dt,
count(event_type) as User_Engagement
FROM events1
WHERE event_type = 'engagement'
group by 1
order by 1
limit 1)

Union 

(SELECT substring(date_trunc('week', occurred_at),1,10) as Activity_dt,
count(event_type) as User_Engagement
FROM events1
WHERE event_type = 'engagement'
group by 1
order by 1 desc
limit 1))
```

#### Question 3

Are changes in user activity/engagement associated with specific devices or hardware ( mobile phone, tablet computers, desktop computers, etc.)  that customers use to read emails and interact with Ziltoâ€™s  web portal?

##### Analysis

From the graph, it can be seen that there is a noticeable dip in the engagement from all the three devices from 28 July to 8 August. Computer devices show a drop of 14.19%, Phone shows 19.68% where as Tablet shows a drop of 29.91%.
Since a significant drop is seen in phones and tablets, it could be due to technical issues with the app such as poorly planned backend, badly designed user interface, third-party user control, app does not support all types of mobile or tablets etc.
Recommendation: Apps can be improved by keeping the interface simple and target right audience. This way the customer retention will be higher which in turn will help Zilto's business.

```
%sql
SELECT DATE_TRUNC('Week', e.occurred_at) AS Week,
       COUNT(DISTINCT e.user_id) AS weekly_users,
       COUNT(DISTINCT CASE WHEN e.device IN ('macbook pro', 'acer aspire notebook','acer aspire desktop','lenovo thinkpad', 'mac mini', 'dell inspiron desktop','dell inspiron notebook','windows surface','macbook air','asus chromebook','hp pavilion desktop') THEN e.user_id ELSE NULL END) AS computer,
       COUNT(DISTINCT CASE WHEN e.device IN ('iphone 5s','nokia lumia 635','amazon fire phone','iphone 4s','htc one','iphone 5','samsung galaxy s4') THEN e.user_id ELSE NULL END) AS phone,
       COUNT(DISTINCT CASE WHEN e.device IN ('kindle fire','samsung galaxy note','ipad mini','nexus 7','nexus 10','samsumg galaxy tablet','nexus 5','ipad air') THEN e.user_id ELSE NULL END) AS tablet
FROM events1 e
JOIN emails em 
on e.user_id = em.user_id
WHERE event_type = 'engagement'
and em.action = 'email_open'
GROUP BY 1
ORDER BY 1
```

#### Question 4
Are changes in user activity/engagement associated with specific countries or regions of the world where Ziltoâ€™s users are located?

##### Analysis

This is the analysis of user engagment with different locations over the week. The data in this line graph is in series grouping of the locations. The line graph represents the week as independent data points and num_engagement as dependent data points to track increase/decrease in user_engagement over the weeks..The highest user engagement on Zilto's portal is in the United States and second highest being Japan. The United States shows a drop of 28.99% from July 28 to Aug 4. It is worth investigating the issue further since the United States has the highest number of zilto users.

```
%sql
SELECT cast(Date_Trunc('Week', e.occurred_at) as date) AS Week, location, count(e.user_id),
  count(1) as num_engagements
  FROM events1 e
  WHERE event_type = 'engagement'
  GROUP BY  location, Week
  ORDER BY  num_engagements desc
```

```
%sql
SELECT DATE_TRUNC('week', occurred_at) AS week,
       COUNT(CASE WHEN em.action = 'sent_weekly_digest' THEN em.user_id ELSE NULL END) AS weekly_digest_emails,
       COUNT(CASE WHEN em.action = 'sent_reengagement_email' THEN em.user_id ELSE NULL END) AS reengagement_emails,
       COUNT(CASE WHEN em.action = 'email_open' THEN em.user_id ELSE NULL END) AS email_opens,
       COUNT(CASE WHEN em.action = 'email_clickthrough' THEN em.user_id ELSE NULL END) AS email_clickthroughs
FROM emails em
GROUP BY 1
ORDER BY 1
```

```
%sql
-- Further investigation on question 4
-- Checking the number of active users
%sql
SELECT DATE_TRUNC('week', e.occurred_at),
       COUNT(DISTINCT e.user_id) AS weekly_active_users
FROM events1 e
WHERE e.event_type = 'engagement'
  AND e.event_name = 'login'
GROUP BY 1
ORDER BY 1 
```

It can be seen that there is a steep drop of 17.46% in active users (i.e. drop in logins) from july 28 to August 4. The location analysis(Q4) shows that the highest number of active users from United States have been reduced during the same time. One of the speculations to this could be the summer season (vacation) in the United States that led to drop in active users.
